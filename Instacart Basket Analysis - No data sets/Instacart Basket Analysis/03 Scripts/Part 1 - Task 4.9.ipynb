## Missing Data

```python
#Check each column and count all nulls
customers.isnull().sum()
```

```python
#View the "First Name" column nulls 
customers['First Name'].value_counts(dropna = False)
```

```python
#view the "First Name" column nulls 
customers_nan = customers[customers['First Name'].isnull() == True]
```

```python
customers_nan.head(100)
```

### Only the "First Name" column has missing values (11259). I did not see a pattern as to why these first names were missing (i.e. different dates, last names, customer ids, genders, states, age, etc.). I think that the first names were just omitted, whether by the customer themselves did it or an Instacart employee did it while signing the customer up. As a solution, I will replace the "NaN" values with "missing". 

```python
#Impute the null values in "First Name"
#df_ords['days_since_prior_order'].fillna(0, inplace=True)
customers['First Name'].fillna('missing', inplace = True)
```
